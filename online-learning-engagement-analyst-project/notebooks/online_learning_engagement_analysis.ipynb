# Online Learning Engagement Analysis

This notebook explores student behavior, engagement, performance, and dropout patterns in an online learning dataset.

## Objectives
- Profile the dataset and validate data quality
- Understand engagement and performance behavior
- Analyze dropout patterns
- Surface business-ready insights for stakeholders


```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/online_learning_engagement_dataset.csv')
df.head()
```

## Summary Statistics

```
df.describe(include='all').transpose().head(18)
```

## Dropout Overview

```
dropout_rate = df['dropout'].mean()
dropout_rate
```

```
dropout_by_country = df.groupby('country', as_index=False)['dropout'].mean().sort_values('dropout', ascending=False)
dropout_by_country
```

## Visual Analysis

```
plt.figure(figsize=(8,5))
plt.hist(df['engagement_score'], bins=30)
plt.title('Distribution of Engagement Score')
plt.xlabel('Engagement Score')
plt.ylabel('Student Count')
plt.tight_layout()
plt.show()
```

```
attendance_bins = pd.qcut(df['attendance_rate'], q=5, duplicates='drop')
attendance_summary = df.groupby(attendance_bins, observed=False)['dropout'].mean()

plt.figure(figsize=(8,5))
plt.plot(range(1, len(attendance_summary)+1), attendance_summary.values, marker='o')
plt.xticks(range(1, len(attendance_summary)+1), [f'Q{i}' for i in range(1,len(attendance_summary)+1)])
plt.title('Dropout Rate Across Attendance Rate Quintiles')
plt.xlabel('Attendance Rate Quintile')
plt.ylabel('Dropout Rate')
plt.tight_layout()
plt.show()
```

```
sample = df.sample(5000, random_state=42)
plt.figure(figsize=(8,5))
plt.scatter(sample['engagement_score'], sample['final_grade'], alpha=0.35)
plt.title('Engagement Score vs Final Grade')
plt.xlabel('Engagement Score')
plt.ylabel('Final Grade')
plt.tight_layout()
plt.show()
```

## Correlations

```
num_cols = [c for c in df.columns if df[c].dtype != 'object']
corr = df[num_cols].corr(numeric_only=True)
corr['dropout'].sort_values(key=lambda s: s.abs(), ascending=False)
```

## Initial Insights

- The dataset contains **50,000 rows** and **18 columns**.
- There are **no missing values** in the dataset.
- Average dropout rate is **32.41%**.
- `attendance_rate` has the strongest relationship with dropout in this dataset, with a correlation of **-0.809**.
- `engagement_score` and `final_grade` have a modest positive relationship of **0.050**.
